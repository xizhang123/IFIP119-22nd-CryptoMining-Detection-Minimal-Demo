```python
import os
import shutil
import numpy as np
from matplotlib import pyplot as plt
```

# 显示挖矿木马启动后的CPU占用情况
# Show the CPU usage curve after the malware is activated

```python
with open("hash.txt","r") as f:
    sha256_hash = [hash.strip() for hash in list(f)]
capture = []
for hash in sha256_hash:
    fname = f"02_cpu_usage/{hash}.npy"
    if os.path.exists(fname):
        cpu_line = np.load(fname)
        # 画出样本启动后的CPU占用曲线（在宿主机记录，虚拟机最高50%）
        # draw the CPU usage curve after the malware is activated(recorded in the host machine, the virtual machine has a maximum CPU usage of 50%)
        plt.plot(cpu_line)
        capture += [f"03_traffic/{hash}.pcap"]
plt.show()
```

# 从pcap数据包中提取元数据
# extract metadata from pcap files

```python
for pcap_file in capture:
    !pcap2meta.exe {pcap_file}
```

# 将元数据移动到指定目录
# move the metadata to the specified directory

```python
for hash in sha256_hash:
    shutil.move(f"{hash}.csv",f"04_csv/{hash}.csv")
```
